```python
import numpy as np

def huber_loss(y, y_hat, delta=1.0):
    error = y - y_hat
    abs_error = np.abs(error)

    loss = np.where(
        abs_error <= delta,
        0.5 * error**2,
        delta * (abs_error - 0.5 * delta)
    )
    return np.mean(loss)

# Example values
y = np.array([2, 4, 6, 8, 20], dtype=float)
y_hat = np.array([1, 4, 7, 10, 13], dtype=float)

print("Huber Loss:", huber_loss(y, y_hat, delta=2.0))
```
