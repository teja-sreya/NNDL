```python
import tensorflow as tf
from tensorflow.keras.layers import TextVectorization


texts = [
    "Congrats! You have won a free lottery ticket. Click the link now!",   # spam
    "You are a winner! Claim your prize urgently at http://scam.com",      # spam
    "Hey, are we still meeting for lunch tomorrow?",                       # ham
    "Don't forget to submit the assignment by tonight.",                   # ham
    "URGENT! Your account has been selected for a cash reward!!!"          # spam
]

# 1 = spam, 0 = ham
labels = [1, 1, 0, 0, 1]

x = tf.constant(texts)
y = tf.constant(labels)


vec = TextVectorization(
    max_tokens=5000,
    output_mode="int",
    output_sequence_length=20
)

vec.adapt(x)
x_vec = vec(x)


vocab_size = len(vec.get_vocabulary())

model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, 8),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

model.compile(optimizer="nadam",
              loss="binary_crossentropy",
              metrics=["accuracy"])


model.fit(x_vec, y, epochs=20, batch_size=2, verbose=1)


loss, acc = model.evaluate(x_vec, y, verbose=0)
print("Accuracy:", acc)

msg = ["You are a winner U have been specially selected to win big!!"]
p = model.predict(vec(msg), verbose=0)[0][0]
print("Spam probability:", float(p))
print("Prediction:", "SPAM" if p >= 0.5 else "HAM")
```
